```python
Nut = {"Almond" : 20 , "Badam" : 31 , "Kaju" : 27 , "Pista" : 21}
fig, ax = plt.subplots()
ax.bar(Nut.keys(),Nut.values())
ax.set(title="COmparing costs of dryfruits",xlabel="Dryfruits",ylabel="Cost")
fig.show()
```

```python
fig , ((ax1, ax2),(ax3, ax4)) = plt.subplots(2,2,figsize=(10,5))
ax1.plot(x, np.array(x)/2)
ax2.scatter(np.random.random(10),np.random.random(10))
ax3.bar(Nut.keys(),Nut.values())
ax4.hist(np.random.random(100))
fig.show()
```

```python
#method 2
fig , ax = plt.subplots(2,2,figsize=(10,5))
ax[0,0].plot(x, np.array(x)/2)
ax[0,1].scatter(np.random.random(10),np.random.random(10))
ax[1,0].bar(Nut.keys(),Nut.values())
ax[1,1].hist(np.random.random(100))
fig.show()
```
